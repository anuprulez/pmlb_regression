```python
import os
import sys
import time
import pandas as pd
from sklearn import model_selection
from sklearn import linear_model
from matplotlib import pyplot as plt
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from sklearn.model_selection import GridSearchCV, KFold
import warnings


warnings.filterwarnings('ignore')
%matplotlib inline

path = "scikit_20/data_used"
datasets = os.listdir(path)

print("Number of datasets: %d" % len(datasets))
```

```python
regressor_no_normalization = Pipeline([
  ('regression', linear_model.ElasticNet())
])

regressor_normalization = Pipeline([
  ('prep', preprocessing.Normalizer()),
  ('regression', linear_model.ElasticNet())
])

parameters = [{}]
r2_scores = list()
r2_scores_normalized = list()

for d_set in datasets:
    data_path = path + "/" + d_set
    df = pd.read_csv(data_path, sep="\t")
    label = df["target"].copy()
    data = df.drop("target", axis=1)
    optimized_regressor_no_normalization = GridSearchCV(regressor_no_normalization, parameters, \
                                       cv=KFold(n_splits=3, shuffle=True, random_state=3111696), \
                                       error_score=0, scoring='r2')
    
    optimized_regressor_normalization = GridSearchCV(regressor_normalization, parameters, \
                                       cv=KFold(n_splits=3, shuffle=True, random_state=3111696), \
                                       error_score=0, scoring='r2')
    optimized_regressor_no_normalization.fit(data, label)
    optimized_regressor_normalization.fit(data, label)

    best_score_no_normalization = optimized_regressor_no_normalization.best_score_
    r2_scores.append(best_score_no_normalization)
    
    best_score_normalization = optimized_regressor_normalization.best_score_
    r2_scores_normalized.append(best_score_normalization)
    
print("Mean R2 square without normalization: %.2f" % np.mean(r2_scores))
print("Mean R2 square with normalization: %.2f" % np.mean(r2_scores_normalized))
print('Training finished')

plt.figure(figsize=(12, 12))
plt.scatter(r2_scores_normalized, r2_scores, color='r')
plt.xlabel('R2 scores with normalization', size=16)
plt.ylabel('R2 scores without normalization', size=16)
plt.title('ElasticNet with and without normalized datasets', size=18)
plt.xticks(size=14)
plt.yticks(size=14)
plt.legend(["R2 scores"], prop={'size': 16})
plt.grid(True)
plt.show()
```
